```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  # Adjust the configuration of the plot we will create

# Read in the data
df = pd.read_csv(r'C:\Users\usa_t\OneDrive\Desktop\Correlation Python Portfolio\movies.csv')
```

```python
# Look at the Data
df.head()
```

```python
# Is there any missing data?
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))

# df = print(df.dropna())
```

```python
# Checking for null values
null_values = df.isnull()
# print(null_values)

# Remove the whole ROWS with null values
df = df.dropna()
# print(df_cleaned)

df
```

```python
# Data Type for our columns
df.dtypes
```

```python
"""
Notes:

i) errors='coerce' 
An argument that specifies how to handle parsing errors. 
When 'coerce' is used, it converts non-numeric values (e.g., strings) to NaN (Not a Number) 
instead of raising an error.

ii).fillna(0):
After applying to_numeric with 'coerce,' some values in the 'budget' column may become NaN. 
This method fills those NaN values with 0, effectively converting them to numeric values.
"""
# Changing data type of budget column to integer and applying it (removing the .0) 
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)

# Changing data type of gross column to integer and applying it 
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int)

# print the dataframe
df

```

```python
"""
Notes:

i) .str.split(): 
This method is applied to each string in the 'released' column, and it splits the string into a list of substrings using whitespace as the delimiter. 
This effectively separates the date into its individual components: month, day, and year.

ii).str[2]: 
After splitting the string, this part selects the third element (index 2) of the resulting list, which corresponds to the year. 

"""

# Extracting the year from released column to be the same as the year column
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

# print the dataframe
df
```

```python
# Sort the values in gross column from highest to lowest 
df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df
```

```python
# Display the full rows and columns
# pd.set_option('display.max_rows', None)
```

```python
# Checking for duplicates 
duplicates = df.duplicated()
# print(duplicates)

# Remove the duplicates 
df = df.drop_duplicates()
# print(df_no_duplicates)
```

```python
# Compare budget with gross rvenue
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross')
plt.show()
```

```python
# Plot the budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"purple"})
```

```python
# Correlation ( Pearson,Kendall,Spearman)

# Exclude non numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Find correlation of only numeric columns
correlations = numeric_df.corr(method='pearson')
print(correlations)
```

```python
# High correlation between budget and gross
correlation_matrix = correlations
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
